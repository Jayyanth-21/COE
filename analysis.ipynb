```python
from scipy.stats import norm
mu = 70
sigma = 10
x = np.linspace(mu-4*sigma,mu+4*sigma,100)
pdf = norm.pdf(x,mu,sigma)
plt.plot(x,pdf,color='r',label=f'N({mu},{sigma})')
plt.xlabel('Exam scores')
plt.ylabel('probability density')
plt.title('Normal distributi9on')
plt.legend()
plt.grid()
```

```python
z=np.linspace(-4,4,1000)
pdf=norm.pdf(z,0,1)

plt.plot(z,pdf,color='green')

plt.xlabel('Zscore')
plt.ylabel("probability density")
plt.axvline(x=0,color='red',linestyle='--',label='Mean (z=0)')
plt.legend()
plt.grid()
plt.show()
```

```python
np.random.seed(42)
income_data=np.random.exponential(scale=50,size=1000)
plt.figure(figsize=(8,4))
sns.histplot(income_data,bins=60,kde=True,color='r')
plt.show()

sample_means=[]
n_samples=1000
sample_size=30

for _ in range(n_samples):
    sample=np.random.choice(income_data,size=sample_size)
    sample_means.append(np.mean(sample)) 
sns.histplot(sample_means,bins=60,kde='True',color='b')
```
