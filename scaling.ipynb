```python
import pandas as pd
import numpy as np

# Load the data
data = pd.read_csv("/content/daytoday.csv")

# Print the column names to check for discrepancies
print("Column names in the DataFrame:")
print(data.columns.tolist())

# Check if the 'LAB DATES' column exists
if 'LAB DATES' in data.columns:
    # Extract the sub-columns under 'LAB DATES'
    lab_dates_columns = data.columns[data.columns.str.contains('LAB DATES')].tolist()
    print("Sub-columns under 'LAB DATES':", lab_dates_columns)
else:
    print("Warning: 'LAB DATES' column does not exist in the DataFrame.")

def convert_date_format(date_str):
    if pd.isna(date_str):
        return date_str
    try:
        # Convert the date string to datetime
        date_obj = pd.to_datetime(date_str, format='%d/%b')
        # Format it to DD-MM-YYYY
        return date_obj.strftime('%d-%m-%Y')  # Change to %Y for four-digit year
    except ValueError:
        # Handle the case where the date format is incorrect
        print(f"Warning: '{date_str}' is not in the expected format and will be returned as NaT.")
        return pd.NaT  # Return Not a Time for invalid dates

# List of date columns to transform (sub-columns under 'LAB DATES')
date_columns = ['15/Jul', '22/Jul', '29/Jul', '05/Aug', '12/Aug', '09/Sep',
                '21/Sep', '23/Sep', '30/Sep', '07/Oct', '14/Oct', '28/Oct']

# Apply the date conversion to each specified sub-column
for col in date_columns:
    full_col_name = f'LAB DATES.{col}'  # Construct the full column name
    if full_col_name in data.columns:  # Check if the full column exists in the DataFrame
        data[full_col_name] = data[full_col_name].apply(convert_date_format)  # Replace with formatted date
    else:
        print(f"Warning: Column '{full_col_name}' does not exist in the DataFrame.")

# Save the transformed data to a new CSV file
output_file_path = 'daytoday_transformed.csv'
data.to_csv(output_file_path, index=False)

print("Date format transformation completed and saved to:", output_file_path)
```

```python
#STANDARD SCALER
import pandas as pd
import numpy as np

df=pd.read_csv("/content/daytodayeval_transf.csv")
tot_val=df.iloc[:,-1]
sum=0
sdsum=0
for x in tot_val:
  sum+=x
mean=sum/len(tot_val)
for y in tot_val:
  sdsum+=((y-mean)*(y-mean))/len(tot_val)
sd=np.sqrt(sdsum)
print(df.iloc[:,-1].std())

print(sd)
for v in range(len(tot_val)):
  df.iloc[v,-1]=(df.iloc[v,-1]-mean)/sd

print(np.mean(df.iloc[:,-1]))
print(np.std(df.iloc[:,-1]))
```
