```python
for i in df:
    if df[i].dtype=='object':
        df[i]=df[i].str.lower()
```

```python
df.info()
```

```python
df['number_of_cylinders']=df['number_of_cylinders'].map({'two':2,'three':3,'four':4,'five':5,'six':6,'eight':8,'twelve':12})
```

```python
df['fuel_system']=df['fuel_system'].map({
    'mpfi': 'Multi-Point Fuel Injection',
    'spdi': 'Single-Point Fuel Injection',
    '2bbl': 'Two-Barrel Carburetor',
    '1bbl': 'One-Barrel Carburetor',
    'idi': 'Indirect Injection',
    'dohc': 'Dual Overhead Cam',
    'ohc': 'Overhead Cam',
    'lpg': 'Liquefied Petroleum Gas',
    'diesel': 'Diesel Engine',
})
```

```python
df['horsepower']=df['horsepower'].astype(int)
```

```python
df['peak_rpm']=df['peak_rpm'].astype(int)
```

```python
df['price']=df['price'].astype(int)
```

```python
df['hp_KW']=df['horsepower']*0.7457
```

```python
df['hp_KW']
```

```python
df['Power_to_weight_ratio']=df['horsepower']/df['curb_weight']
```

```python
df['Engine_eff']=df['horsepower']/df['engine_size']
```

```python
df['Fuel_eff']=df['city_mpg']/df['highway_mpg']
```

```python
import numpy as np
```

```python
df['categories']=np.where(df['compression_ratio']<9,'low',
                 np.where(df['compression_ratio']<=11,'medium','high'))
df
```
